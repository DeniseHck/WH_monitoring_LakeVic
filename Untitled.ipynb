```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

OA = pd.read_csv("OA.csv")
#print(OA)

OA['Date'] = pd.to_datetime(OA['Date'], format = '%d/%m/%Y')
#print(OA)

#xaxis = OA["Date"]
#yaxis = OA["RF"]

#plt.bar(xaxis,yaxis)

# set width of bar
barWidth = 0.35
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
IT = OA["RF"]
ECE = OA["THR"]
text = OA["Date"]

meanRF = IT.mean()
meanTHR = ECE.mean()

print("The mean OA for RF is", meanRF)
print("The mean OA for THR is", meanTHR)

#code to remove the seconds, hours and minutes
app = []
for i in text:
    a = i.strftime("%Y-%m-%d")
    app.append(a)
#print(app)
 
# Set position of bar on X axis
br1 = np.arange(len(IT))

br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
# Make the plot
plt.bar(br1, IT, color ='mediumturquoise', width = barWidth,
        edgecolor ='grey', label ='Random Forest')
plt.bar(br2, ECE, color ='teal', width = barWidth,
        edgecolor ='grey', label ='Thresholding')
#plt.bar(br3, CSE, color ='b', width = barWidth,
        #edgecolor ='grey', label ='CSE')
 
#Adding Xticks
plt.xlabel('Image date', fontsize = 12)
plt.ylabel('Overall Accuracy', fontsize = 12)
plt.xticks([r + barWidth for r in range(len(IT))],
        app)
plt.ylim([0,1.1])
 
plt.legend()
plt.xticks(rotation=90)
plt.savefig('../Desktop/Figgys/OA.pdf')
```
