```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns

WH_spectra = pd.read_csv("WH_spectra.csv")

#S1['Month step'] = pd.to_datetime(S1['Month step'], format = '%d/%m/%Y')

#S1[S1==0] = np.nan
#sentinel1=S1.dropna()

#WH_spectra

#switch rows and columns
WH_spectra_tr = WH_spectra.transpose()

#reverse row index (smaller wavelengths first)
WH_spectra_tr_rev = WH_spectra_tr[::-1]

#print(WH_spectra_tr_rev)
#plt.plot(WH_spectra_tr_rev)
df = WH_spectra_tr_rev.drop("AOT").drop("WVP").drop("SCL")
#print(df)

#plt.plot(df)

x = df[0]

#print(x)




df2 = df.loc[:, 1:90:1]
#print(df2)

df2['mean_rows'] = df2.mean(axis = 1)
#print(df2)
df2["min_rows"] = df2.min(axis=1)
df2["max_rows"] = df2.max(axis=1)

y = df2['mean_rows']#.iloc[::-1]
y1 = df2["min_rows"]#
y2 = df2["max_rows"]
#print(y)
#print(y1)



plt.figure(figsize=(17,4))
#plt.plot(x,y, color="purple")
#plt.plot (x, y1, color="blue")
#plt.plot(x, y2, color="blue")
plt.fill_between(x, y2,y1, color="violet", alpha =0.2)
plt.plot(x,y)
#no AOT, no WVP, no SCL

```

```python
#CY_spectra_2018.csv

CY_spectra = pd.read_csv("CY_spectra.csv")

#switch rows and columns
CY_spectra_tr = CY_spectra.transpose()

#reverse row index (smaller wavelengths first)
CY_spectra_tr_rev = CY_spectra_tr[::-1]

#print(WH_spectra_tr_rev)
#plt.plot(WH_spectra_tr_rev)
df_CY = CY_spectra_tr_rev.drop("AOT").drop("WVP").drop("SCL")
#print(df)

#plt.plot(df)

x_CY = df_CY[0]

#print(x_CY)



df2_CY = df_CY.loc[:, 1:101:1]
#print(df2_CY)

df2_CY['mean_rows'] = df2_CY.mean(axis = 1)
#print(df2_CY)

df2_CY["min_rows"] = df2_CY.min(axis=1)
df2_CY["max_rows"] = df2_CY.max(axis=1)

y_CY = df2_CY['mean_rows']#.iloc[::-1]
y1_CY = df2_CY["min_rows"]#
y2_CY = df2_CY["max_rows"]
#print(y_CY)
#print(y1_CY)


plt.figure(figsize=(17,4))
#plt.plot(x_CY,y_CY, color="purple")
#plt.plot (x, y1, color="blue")
#plt.plot(x, y2, color="blue")
plt.fill_between(x_CY, y2_CY, y1_CY, color="green", alpha =0.2)
plt.plot(x_CY,y_CY, color="green")
#no AOT, no WVP, no SCL

```
